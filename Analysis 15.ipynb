```python
import csv
import pandas as pd
import matplotlib as plt
% matplotlib inline
```

```python
df = pd.read_csv("Analyses/Final Data.csv", encoding = 'iso-8859-1')
df = df[df['Include in Data Analysis'] == 1]
```

```python
buildings = df['Primary Property Type Category - Reduced']
```

```python
df = df.drop('Primary Property Type Category - Reduced', 1)
```

```python
df = df.apply(pd.to_numeric, errors = 'coerce')
```

```python
df['Primary Property Type Category - Reduced'] = buildings
```

```python
sums_energy_use_by_building_type = df[['Electricity Use - Grid Purchase and Generated from Onsite Renewable Systems (kBtu)',\
                                      'Natural Gas Use (kBtu)','Fuel Oil #1 Use (kBtu)','Fuel Oil #2 Use (kBtu)',\
                                      'Fuel Oil #4 Use (kBtu)', 'Fuel Oil #5 & 6 Use (kBtu)', 'Diesel #2 Use (kBtu)',\
                                       'Kerosene Use (kBtu)', 'Propane Use (kBtu)', 'District Steam Use (kBtu)', 'District Hot Water Use (kBtu)',\
                                       'District Chilled Water Use (kBtu)', 'Coal - Bituminous Use (kBtu)', 'Coal - Anthracite Use (kBtu)',\
                                       'Other Use (kBtu)', 'Wood Use (kBtu)', 'Coke Use (kBtu)', 'Site Energy Use (kBtu)',\
                                       'Primary Property Type Category - Reduced']].groupby('Primary Property Type Category - Reduced').sum()
```

```python
sums_energy_use_by_building_type['electricity_perc'] = sums_energy_use_by_building_type['Electricity Use - Grid Purchase and Generated from Onsite Renewable Systems (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['nat_gas_perc'] = sums_energy_use_by_building_type['Natural Gas Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['fuel_1_perc'] = sums_energy_use_by_building_type['Fuel Oil #1 Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['fuel_2_perc'] = sums_energy_use_by_building_type['Fuel Oil #2 Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['fuel_4_perc'] = sums_energy_use_by_building_type['Fuel Oil #4 Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['fuel_5&6_perc'] = sums_energy_use_by_building_type['Fuel Oil #5 & 6 Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['diesel_2_perc'] = sums_energy_use_by_building_type['Diesel #2 Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['kerosene_perc'] = sums_energy_use_by_building_type['Kerosene Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['propane_perc'] = sums_energy_use_by_building_type['Propane Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['dist_steam_perc'] = sums_energy_use_by_building_type['District Steam Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['dist_hot_water_perc'] = sums_energy_use_by_building_type['District Hot Water Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['dist_chilled_water_perc'] = sums_energy_use_by_building_type['District Chilled Water Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['coal_bit_perc'] = sums_energy_use_by_building_type['Coal - Bituminous Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['coal_anth_perc'] = sums_energy_use_by_building_type['Coal - Anthracite Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['other_perc'] = sums_energy_use_by_building_type['Other Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['wood_perc'] = sums_energy_use_by_building_type['Wood Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
sums_energy_use_by_building_type['coke_perc'] = sums_energy_use_by_building_type['Coke Use (kBtu)'] / sums_energy_use_by_building_type['Site Energy Use (kBtu)']
```
