```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv("Analyses/Final Data.csv", encoding='iso-8859-1')
#Limiting to rows that we want to include (Include == 1)
df = df[df['Include in Data Analysis'] == 1]
df['Percent of Floor Area Occupied by Additional Space Use Types'] = df['% of Floor Area Occupied by Additional Space Use Types (Percentage)']
```

```python
df = df.apply(pd.to_numeric, errors = 'coerce')
```

```python
#14 Median site and source EUI by secondary space use
site_source_EU_vs_secondary = df[[
        'Weather Normalized Source EUI (kBtu/ft2)', 
        'Weather Normalized Site EUI (kBtu/ft2)', 
        'Percent of Floor Area Occupied by Additional Space Use Types']]
site_source_EU_vs_secondary.to_csv("14 - EUI vs secondary space use.csv")
```

```python
# plt.figure(figsize = (12,4))
ax = df.plot.scatter(x = 'Percent of Floor Area Occupied by Additional Space Use Types', y = 'Weather Normalized Source EUI (kBtu/ft2)', color = 'c', label = 'WN Source EUI', figsize = (12,8))
fig = df.plot.scatter(x = 'Percent of Floor Area Occupied by Additional Space Use Types', y = 'Weather Normalized Site EUI (kBtu/ft2)', color = 'Red', label = 'WN Site EUI', ax=ax)
plt.xlim([0,70])
plt.ylim([0,1500])
plt.title('EUI vs. Secondary Space Use', size = 20)
plt.savefig('scatter.png')
```
